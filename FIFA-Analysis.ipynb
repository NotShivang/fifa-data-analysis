```python
# remove the currency symbol and letters M and K
# if value is in thousands, then divide to render all the values in terms of millions

def fixMoney(x):
    try:
        x = x.replace('€','')
        lastChar = len(x)-1
        thousand = False
        if x[lastChar] == 'K':
            thousand = True
            x = x.replace('K','')
        elif x[lastChar] == 'M':
            thousand = False
            x = x.replace('M','')
        x = float(x)
        if thousand:
            x = x/1000
    except:
        x = 0
    return x


fifa['Value'] = fifa['Value'].apply(fixMoney)
```
