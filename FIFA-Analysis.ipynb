```python
# remove the currency symbol and letters M and K
# if value is in thousands, then divide to render all the values in terms of millions

def fixMoney(x):
    try:
        x = x.replace('€','')
        lastChar = len(x)-1
        thousand = False
        if x[lastChar] == 'K':
            thousand = True
            x = x.replace('K','')
        elif x[lastChar] == 'M':
            thousand = False
            x = x.replace('M','')
        x = float(x)
        if thousand:
            x = x/1000
    except:
        x = 0
    return x
```

```python
## minimize the budget
## maximize the overall rating

def bestEleven(budget):
    team = []
    players = 0
    defenders = 0
    midfielders = 0
    forwards = 0
    emptyNet = True
    
    defendersCost = 0
    midfieldersCost = 0
    forwardsCost = 0
    
    teamCost = 0
    teamRating = 0
    
    for i in range(len(fifa_clean1[0:1000])):
        
        forwardsBudget = budget/3
        midfieldersBudget = budget/3
        defendersBudget = budget/4
        keeperBudget = budget - (forwardsBudget + midfieldersBudget + defendersBudget)
           
        ## calculate total cost of player
        value = fifa_xi.loc[i, 'Value']
        wage = fifa_xi.loc[i, 'Wage']

        playerCost = value + 52*wage
        teamCost += playerCost
        
        # position
        playerPos = fifa_xi.loc[i, 'Position']
        
        if players <= 11:
            if playerPos == 'D' and defenders <= 4:
                defendersCost += playerCost
                if defendersCost <= defendersBudget: 
                    team.append(fifa_xi.loc[i,'Name'])
                    players += 1
                    defenders += 1
                    defendersBudget -= playerCost
            elif playerPos == 'M' and midfielders <= 3:
                midfieldersCost += playerCost
                if midfieldersCost <= midfieldersBudget: 
                    team.append(fifa_xi.loc[i,'Name'])
                    players += 1
                    midfielders += 1
            elif playerPos == 'F' and forwards <= 3:
                forwardsCost += playerCost
                if forwardsCost <= forwardsBudget: 
                    team.append(fifa_xi.loc[i,'Name'])
                    players += 1
                    forwards += 1
            else: ## keeper
                if playerCost <= keeperBudget and emptyNet:
                    team.append(fifa_xi.loc[i,'Name'])
                    players += 1
                    emptyNet = False
    
    print(team)
```
